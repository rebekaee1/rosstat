```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('output/ipc_monthly.csv', parse_dates=['date'], index_col='date')
df = df.rename(columns={'ipc': 'value'})
df
```

```python
import statsmodels.api as sm
from dateutil.relativedelta import relativedelta

def train_ols_forecast(data, forecast_steps=12):
  forecast_horizon = forecast_steps
  window_size = len(data)
  monthly_dates = []
  for i in range(12):
      next_date = data.index[-1] + relativedelta(months=i)
      monthly_dates.append(next_date)
  actuals = []
  for i in range(len(data) - forecast_steps):
    actuals.append(np.prod(data.iloc[i - (12 - forecast_horizon): i + forecast_horizon, 0] + 1) * 100 - 100)
  actuals = pd.DataFrame(actuals, index = data.index[-len(actuals):])
  forecast = []
  for i in range(len(data), len(data) +1):
    forecasts_aux = []
    trivial_aux = []
    for m in range(1, forecast_horizon + 1):
      forc = []
      var = []
      for k in range(1, 5):
          p_max = 0.01
          df_aux = data.iloc[i - window_size // k:i, 0].copy()
          # df_aux = df_aux.rolling(m).median()
          while np.sum(np.abs(df_aux - np.mean(df_aux)) / np.std(df_aux) > 3) > 0:
              df_aux[(df_aux - np.mean(df_aux)) / np.std(df_aux) > 3] = np.mean(df_aux) + 1.9 * np.std(df_aux)
              df_aux[(df_aux - np.mean(df_aux)) / np.std(df_aux) < -3] = np.mean(df_aux) - 1.9 * np.std(df_aux)
          train = pd.DataFrame(df_aux)
          # train = pd.DataFrame(df_aux.rolling(forecast_horizon).mean())
          lags_n = 12 + m #if window_size // k > 40 else window_size // k // 7
          if m <= 9:
              lags = [m, m + 1, m + 2, 12]
          elif m == 10:
              lags = [m, m + 1, 12]
          elif m == 11:
              lags = [m, 12]
          elif m == 12:
              lags = [12]
          X_p = [1]
          lags = range(m, m + 3)
          for j in lags:
              train['y_lag' + str(j)] = df_aux.shift(j)
              X_p.append(list(df_aux)[-j + m - 1])
          X_p = pd.DataFrame(np.array(X_p).reshape(1, -1), columns = train.columns)
          spisok = []
          matrix = train.corr()
          cor_max = 0.7
          target = 'value'
          for k in matrix.columns:
              for p in matrix.index:
                  if k != p and k != target and p != target:
                      if np.abs(matrix[k][p]) > cor_max:
                          if np.abs(matrix[target][p]) > np.abs(matrix[target][k]):
                              spisok.append(k)
                          else:
                              spisok.append(p)
          spisok = set(spisok)
          train = train.drop(spisok, axis = 1)

          X_p = X_p.drop(spisok, axis = 1)
          X_p = list(X_p.loc[0])

          train = train.dropna()
          model = sm.OLS(train['value'], sm.add_constant(train.drop('value', axis = 1))).fit()
          X = train.drop('value', axis = 1)
          while np.max(model.pvalues[1:]) > p_max:
              X = X.drop(X.columns[np.argmax(model.pvalues[1:])], axis = 1)
              X_p.pop(np.argmax(model.pvalues[1:]) + 1)
              model = sm.OLS(train['value'], sm.add_constant(X)).fit()

          forc.append(model.predict(X_p)[0])
          var.append(model.mse_resid)
      forecasts_aux.append(np.sum(np.array(forc) / np.array(var)) / np.sum(1 / np.array(var)))
      forecast.append(np.prod(data.iloc[i - (12 - m): i, 0] + 1) * np.prod([x + 1 for x in forecasts_aux]) * 100 - 100)
  forecast = pd.DataFrame(forecast, index = monthly_dates)

  return forecast, actuals

data = df.copy() / 100 - 1

forc, act = train_ols_forecast(data, forecast_steps=12)
forc
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Исходный ряд ИПЦ
axes[0].plot(df.index, df['value'], color='#2563eb', linewidth=1.2)
axes[0].axhline(y=100, color='gray', linestyle='--', alpha=0.5)
axes[0].set_title('ИПЦ (месяц к предыдущему месяцу), %', fontsize=14)
axes[0].set_ylabel('ИПЦ, %')
axes[0].grid(alpha=0.3)

# Кумулятивный прогноз vs фактические
axes[1].plot(act.index, act.values, color='#2563eb', linewidth=1.2, label='Факт (кумулятивный 12 мес.)')
axes[1].plot(forc.index, forc.values, color='#ef4444', linewidth=2, marker='o', markersize=5, label='Прогноз (кумулятивный 12 мес.)')
axes[1].axhline(y=0, color='gray', linestyle='--', alpha=0.5)
axes[1].set_title('Кумулятивный 12-месячный прогноз ИПЦ (OLS)', fontsize=14)
axes[1].set_ylabel('Накопленный рост, %')
axes[1].legend(fontsize=11)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nПрогноз кумулятивной инфляции на 12 месяцев вперёд:")
print(forc.rename(columns={0: 'Прогноз, %'}).round(2))
```
